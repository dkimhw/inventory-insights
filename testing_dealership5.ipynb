```python
def parse_attr(soup, attr_name):
    attrs = []
    for elm in soup.find_all('span'):
        if attr_name in elm.attrs:
            if elm.attrs[attr_name] != '':
                attrs.append(elm.attrs[attr_name])
    return attrs
```
